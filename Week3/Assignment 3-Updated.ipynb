```python
df9= gpd.read_file('CollisionsZIP.zip', encoding='utf-8', crs='epsg:4326')
```

```python
df9.dtypes
```

```python
fig, ax = plt.subplots(figsize=(10,10))

df9.plot(ax=ax,
                 column='distance',
                 legend=True, 
                 scheme='NaturalBreaks')
# basemap from carto that has a dark background (easier to see)
ctx.add_basemap(ax=ax,
                source=ctx.providers.OpenStreetMap.Mapnik
,
                alpha=1,
                crs=df9.crs.to_string()

                # add transparency to make it less dominant
                )
```

```python
df9.info(verbose=True, show_counts=True)
```

```python
# calculate the bounding box 
bbox = df9.total_bounds

# print the bounding box
print("Bounding Box:")
print("min Longitude:", bbox[0])
print("min Latitude:", bbox[1])
print("max Longitude:", bbox[2])
print("max Latitude:", bbox[3])
```
